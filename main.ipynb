```python
from cky import *
from probabilistic_earley import *
import warnings
warnings.filterwarnings('ignore')
```

## CKY Testing

### NP/Z

```python
sentence = "because the nurse examined the patient recovered".split()
grammar_filename = "pcfg/NPZ_PCFG.txt"
grammar = file2grammar(grammar_filename)
chart, tree, prob = parse_sentence("because the nurse examined the patient recovered", "pcfg/NPZ_PCFG.txt")
pretty_print_chart(chart)

chart, prefix = cky_incremental2(sentence, grammar_filename)
print("prefix probs :", prefix)
print("surprisals   :", surprisal_list(prefix))
fig = plot_surprisal_cky(sentence, surprisal_list(prefix))
fig.savefig("plots/NPZ/CKY/garden.jpg", format="jpg", dpi=300)
fig.show()


sentence = "because the nurse examined , the patient recovered".split()
chart, prefix = cky_incremental2(sentence, grammar_filename)
print("prefix probs :", prefix)
print("surprisals   :", surprisal_list(prefix))
fig = plot_surprisal_cky(sentence, surprisal_list(prefix))
fig.savefig("plots/NPZ/CKY/normal.jpg", format="jpg", dpi=300)
fig.show()

sentence = "the nurse examined the patient".split()
chart, prefix = cky_incremental2(sentence, grammar_filename)
print("prefix probs :", prefix)
print("surprisals   :", surprisal_list(prefix))
fig = plot_surprisal_cky(sentence, surprisal_list(prefix))
fig.savefig("plots/NPZ/CKY/simple1.jpg", format="jpg", dpi=300)
fig.show()

sentence = "the patient recovered".split()
chart, prefix = cky_incremental2(sentence, grammar_filename)
print("prefix probs :", prefix)
print("surprisals   :", surprisal_list(prefix))
fig = plot_surprisal_cky(sentence, surprisal_list(prefix))
fig.savefig("plots/NPZ/CKY/simple2.jpg", format="jpg", dpi=300)
fig.show()
```

## Earley Testing

### NP/Z

```python
pcfg = pcfg_file_to_dict("pcfg/NPZ_PCFG.txt")

chart, prefix_probs = earley("because the nurse examined the patient recovered", pcfg)
surprisals_list = surprisals(prefix_probs)
print(surprisals_list)
sentence = "because the nurse examined the patient recovered"
fig = plot_surprisal_earley(sentence.split(), surprisals_list)
fig.savefig("plots/NPZ/garden.jpg", format="jpg", dpi=300)
fig.show()

chart, prefix_probs = earley("because the nurse examined , the patient recovered", pcfg)
surprisals_list = surprisals(prefix_probs)
print(surprisals_list)
sentence = "because the nurse examined , the patient recovered"
fig = plot_surprisal_earley(sentence.split(), surprisals_list)
fig.savefig("plots/NPZ/normal.jpg", format="jpg", dpi=300)
fig.show()

chart, prefix_probs = earley("the nurse examined the patient", pcfg)
surprisals_list = surprisals(prefix_probs)
print(surprisals_list)
sentence = "the nurse examined the patient"
fig = plot_surprisal_earley(sentence.split(), surprisals_list)
fig.savefig("plots/NPZ/simple1.jpg", format="jpg", dpi=300)
fig.show()

chart, prefix_probs = earley("the patient recovered", pcfg)
surprisals_list = surprisals(prefix_probs)
print(surprisals_list)
sentence = "the patient recovered"
fig = plot_surprisal_earley(sentence.split(), surprisals_list)
fig.savefig("plots/NPZ/simple2.jpg", format="jpg", dpi=300)
fig.show()
```
